```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy="mean")
columns=['CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term',
       'Credit_History']
for i in columns:
    dataset[i] = si.fit_transform(dataset[[i]])
```
