```python
from sklearn.pipeline import Pipeline
#feature scaling 
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
```

```python
steps= [("standard_scaler",StandardScaler()),
        ("classifier",LogisticRegression())]
```

```python
steps
```

```python
pipe=Pipeline(steps)
pipe
```

```python
#visualize pipeline
from sklearn import set_config
```

```python
set_config(display="diagram")
```

```python
pipe
```

```python
##creating a dataset 
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000)
```

```python
X.shape
```

```python
from sklearn.model_selection  import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.33,random_state=42)
```

```python
X_train
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

## Example
Complex examples of columns Transformer

```python
from sklearn.impute import SimpleImputer
```

```python
#numerical processing pipeline
import numpy as np
numeric_processor = Pipeline(
    steps=[("imutation_mean",SimpleImputer(missing_values=np.nan,strategy="mean")),
          ("scaler",StandardScaler()),]
)
```

```python
numeric_processor
```

```python
##categorical procesing pipeline
from sklearn.preprocessing import OneHotEncoder
categorical_processor=Pipeline(
    steps=[("imputation_constant",SimpleImputer(fill_value="missing",strategy="constant")),
          ("onehot",OneHotEncoder(handle_unknown="ignore"))]
)
```

```python
categorical_processor
```

```python
##combine processing techniques
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(
    [("categorical",categorical_processor,["gender","City"]),
    ("numerical",numeric_processor,["age","height"])]
)
```

```python
preprocessor
```

```python
from sklearn.pipeline import make_pipeline
```

```python
final_pipe=make_pipeline(preprocessor,LogisticRegression())
```
