```python
def outlier(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
outlier(df['Your Last Year Percentage'])
```

```python
lw,uw = outlier(df['Your Last Year Percentage'])
df[i]=np.where(df['Your Last Year Percentage']<lw,lw,df['Your Last Year Percentage'])
df[i]=np.where(df['Your Last Year Percentage']>uw,uw,df['Your Last Year Percentage'])
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
columns = ['Gender ','Education Level','Feeling down, depressed, or hopeless ',
       'Little interest or pleasure in doing things ',
       'Are you having proper sleep or not minimum 6hrs daily',
       'Feeling tired or having little energy', 'Taking Proper Meal or Not',
       'Feeling bad about yourself or that you are a failure or not have let yourself or your family down',
       'Did you feel trouble on concentrating on things, such as reading the newspaper or studying or reading books or while watching television and etc...',
       'Moving or speaking so slowly that other people could have noticed Or being so restless that you have been moving around a lot more than usual',
       'Thoughts that you would be better off dead or of hurting yourself in some way',
       'Did you live with your family or not',
       'Do you have part-time or full-time job? ','Do you consume substances such as alcohol, tobacco, or drugs?  ']
for column in columns:
    df[column] = label_encoder.fit_transform(df[column])
    df[column] = label_encoder.fit_transform(df[column].astype('int64'))
    
```

```python
columns = ['How many hours do you spend studying each day(Self Studying)?',
       'How many hour do you spend electronic gadgets (e.g. mobile phone, computer, laptop, PSP, PS4.0, Wii, etc.) ',
       'How many hours do you spend on social media per day?',
       '  How many hours of physical activity or exercise do you engage in weekly?  ']
for column in columns:
    df[column] = df[column].astype('int64')
```

```python
# Define the scoring function considering all columns
def mental_health_score(row):
    score = 0
    
    # High severity indicators
    if row['Thoughts that you would be better off dead or of hurting yourself in some way'] == 1:
        score += 3
    if row['Feeling down, depressed, or hopeless '] == 1:
        score += 3
    if row['Little interest or pleasure in doing things '] == 1:
        score += 3
    
    # Medium severity indicators
    if row['Are you having proper sleep or not minimum 6hrs daily'] == 0:
        score += 2
    if row['Feeling tired or having little energy'] == 1:
        score += 2
    if row['Feeling bad about yourself or that you are a failure or not have let yourself or your family down'] == 1:
        score += 2
    if row['Did you feel trouble on concentrating on things, such as reading the newspaper or studying or reading books or while watching television and etc...'] == 1:
        score += 2
    if row['Moving or speaking so slowly that other people could have noticed Or being so restless that you have been moving around a lot more than usual']== 1:
        score += 2
    
    # Lifestyle and behavior
    if row['Do you consume substances such as alcohol, tobacco, or drugs?  '] == 1:
        score += 1
    if row['How many hours do you spend studying each day(Self Studying)?'] < 1:
        score += 1
    if row['How many hour do you spend electronic gadgets (e.g. mobile phone, computer, laptop, PSP, PS4.0, Wii, etc.) '] > 4:
        score += 1
    if row['How many hours do you spend on social media per day?'] > 4:
        score += 1
    if row['  How many hours of physical activity or exercise do you engage in weekly?  '] < 1:
        score += 1
    
    # Demographic factors
    if row['Did you live with your family or not'] == 0:
        score += 1
    if row['Do you have part-time or full-time job? '] == 0:
        score += 1
    if row['Your Last Year Percentage'] < 50:
        score += 1  # Lower performance may be indicative of stress

    return score

# Apply the function to calculate the total mental health score
df['Mental Health Score'] = df.apply(mental_health_score, axis=1)

# Categorize based on the score
def categorize_by_score(score):
    if score >= 15:
        return 'Severely Affected'
    elif score >= 8:
        return 'Moderately Affected'
    else:
        return 'Healthy'

# Apply the function to categorize based on the score
df['Mental Health Category'] = df['Mental Health Score'].apply(categorize_by_score)

# Check the distribution of the new labels
df['Mental Health Category'].value_counts()
```
